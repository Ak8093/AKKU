```python
# Install Hugging Face Transformers library
!pip install transformers

# Import pipeline
from transformers import pipeline

# -----------------------------
# Experiment 1-a: Sentiment Analysis
# -----------------------------

# Create sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

# Input text
text_input = "I really love the new design of your app!"

# Perform sentiment analysis
sentiment_result = sentiment_analyzer(text_input)

# Display results
print("=== Sentiment Analysis ===")
print(f"Text: {text_input}")
print(f"Sentiment: {sentiment_result[0]['label']} (Score: {sentiment_result[0]['score']:.2f})\n")


# -----------------------------
# Experiment 1-b: Language Translation (English → French)
# -----------------------------

# Create translation pipeline (English to French)
translator_en_to_fr = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")

# Input text
text_input = "I really love the new design of your app!"

# Perform translation
translation_result = translator_en_to_fr(text_input)

# Extract translated text
translated_text = translation_result[0]['translation_text']

# Display results
print("=== Translation (English → French) ===")
print(f"Original: {text_input}")
print(f"French: {translated_text}")
```
